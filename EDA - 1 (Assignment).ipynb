### Missing Values

```python
df['LB'] = df['LB'].fillna(df['LB'].mean())
df['AC'] = df['AC'].fillna(df['AC'].mean())
df['DS'] = df['DS'].fillna(df['DS'].mean())
df['DP'] = df['DP'].fillna(df['DP'].mean())
df['MLTV'] = df['MLTV'].fillna(df['MLTV'].mean())
df['ASTV'] = df['ASTV'].fillna(df['ASTV'].mean())
df['Width'] = df['Width'].fillna(df['Width'].mean())
df['Tendency'] = df['Tendency'].fillna(df['Tendency'].mean())
df['NSP'] = df['NSP'].fillna(df['NSP'].mean())
```

### outlier

```python
Q1 = df['AC'].quantile(0.25)
Q3 = df['AC'].quantile(0.75)
IQR = Q3 - Q1

lo_bond = Q1 - (1.5 * IQR)
up_bond = Q3 + (1.5 * IQR)

df_no_outliers = df[(df['AC']> lo_bond) & (df['AC']> up_bond)]
df_no_outliers.boxplot('AC')
plt.show()
```

### Pattern Recognition and insights:

```python
# Correlatio
sns.heatmap(data=df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation using Heatmap')
plt.show()
```

In the heatmap:
The correlation values range from -1 to 1.
Values closer to 1 indicate a strong positive correlation.
Values closer to -1 indicate a strong negative correlation.
Values around 0 indicate no correlation.
This visualization will help's us identify the relationships between different variables in your dataset.

```python
# Temporal data
df['LB']= pd.to_datetime(df['LB'])
df.sort_values('LB')
x= df['LB']
y= df['ASTV']
plt.plot_date(x,y,linestyle= 'solid')
plt.show()
```
